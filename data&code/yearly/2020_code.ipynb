```python
# Append, sort and check unique labels
import pandas as pd

df_2019 = pd.read_excel('bgg/2019_data.xlsx', index_col=None)
df_2020 = pd.read_excel('bgg/2020_data.xlsx', index_col=None)

pd_2019_20 = pd.concat([df_2020, df_2019])
sorted_df = pd_2019_20.sort_values(by=['game'], ascending=True)

print(sorted_df['game'].unique())
```

```python
# Convert to string prior to merge (otherwise integers get messed up for printing)
t1 = df_2019.astype(str)
t2 = df_2020.astype(str)

# Merge data
merged_df = pd.merge(t2, t1, on='game', how='left')

# Show the resulting
print(merged_df)
```

```python
# Define a function to calculate the necessary padding spaces
def add_spaces(info, desired_length=8):
    # Calculate the number of spaces needed to reach the desired length
    return " " * max(0, desired_length - len(str(info)))

#Write txt version of geeklist
with open("bgg/2020_txt.txt", "a") as file:
    for x in range(0, 125):
        temp = merged_df.iloc[x]['placement_y']

        # Extract values for easier readability
        game = merged_df.iloc[x]['game']
        placement_x = int(merged_df.iloc[x]['placement_x'])
        score_f_x = str(merged_df.iloc[x]['score_f_x'])
        avg_vote_x = str(merged_df.iloc[x]['avg_vote_x'])
        count_tot_x = str(merged_df.iloc[x]['count_tot_x'])
        max_vote_x = str(merged_df.iloc[x]['max_vote_x'])
        count1_x = str(merged_df.iloc[x]['count1_x'])
        note_x = str(merged_df.iloc[x]['note_x'])

        # Prepare note line if applicable
        note_line = ""
        if note_x != ".":
            note_line = "\n"+"Note: "+note_x+"\n"
        
        # if new on list
        if pd.isnull(temp):
            file.write("-----"+merged_df.iloc[x]['game']+"-----"+"\n" # Header: Game
                       +"[b][size=18]#"+str(placement_x)+" [color=#ff2600]**NEW**[/color]"+"[/size][/b]"+"\n" # Row 1
                       +"[c]"+"\n" # Open table
                       +"              [u]2020[/u]    [u]2019[/u]"+"\n" # table header
                       +"Rank:         "+str(placement_x)+"\n"
                       +"Score:        "+score_f_x+"\n"
                       +"Avg. vote:    "+avg_vote_x+"\n"
                       +"# of votes:   "+count_tot_x+"\n"
                       +"Highest vote: "+max_vote_x+"\n"
                       +"#1 votes:     "+count1_x+"\n"
                       +"[/c]"+"\n" # Close table
                       +note_line
                       +"-----"+"\n"# End of entry
                      )
        # if not new on list
        else:
            placement_y = int(merged_df.iloc[x]['placement_y'])
            score_f_y = str(merged_df.iloc[x]['score_f_y'])
            avg_vote_y = str(merged_df.iloc[x]['avg_vote_y'])
            count_tot_y = str(merged_df.iloc[x]['count_tot_y'])
            max_vote_y = str(merged_df.iloc[x]['max_vote_y'])
            count1_y = str(merged_df.iloc[x]['count1_y'])
            
            # Determine arrow and difference if applicable
            if placement_x == placement_y:
                movement = " :arrowE::arrowW:"
            elif placement_x > placement_y:
                difference = placement_x - placement_y
                movement = f" :arrowS: {difference}"
            else:
                difference = placement_y - placement_x
                movement = f" :arrowN: {difference}"
            file.write("-----"+merged_df.iloc[x]['game']+"-----"+"\n" # Header: Game
                       +"[b][size=18]#"+str(placement_x)+"[/size][/b]"+movement+"\n" # Row 1
                       +"[c]"+"\n" # Open table
                       +"              [u]2020[/u]    [u]2019[/u]"+"\n" # table header
                       +"Rank:         "+str(placement_x)+add_spaces(str(placement_x))+str(placement_y)+"\n"
                       +"Score:        "+score_f_x+add_spaces(score_f_x)+score_f_y+"\n"
                       +"Avg. vote:    "+avg_vote_x+add_spaces(avg_vote_x)+avg_vote_y+"\n"
                       +"# of votes:   "+count_tot_x+add_spaces(count_tot_x)+count_tot_y+"\n"
                       +"Highest vote: "+max_vote_x+add_spaces(max_vote_x)+max_vote_y+"\n"
                       +"#1 votes:     "+count1_x+add_spaces(count1_x)+count1_y+"\n"
                       +"[/c]"+"\n" # Close table
                       +note_line
                       +"-----"+"\n"# End of entry
                      )
```
